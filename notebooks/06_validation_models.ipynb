```python
best_period = 3.2886
best_duration = 0.1400
best_depth = 0.002298
```

```python
tic = Catalogs.query_criteria(
    catalog="Tic",
    ID=25155310
)

tic
```

```python
stellar_radius = tic["rad"][0]
effective_temp = tic["Teff"][0]
contamination = tic["contratio"][0]

print("Stellar radius:", stellar_radius)
print("Temperature:", effective_temp)
print("Contamination ratio:", contamination)
```

```python
planet_radius = stellar_radius * np.sqrt(best_depth)

print("Estimated planet radius:", planet_radius)
```

```python
planet_radius_rj = planet_radius * 9.73

print(f"{planet_radius_rj:.2f} Jupiter radii")
```

```python
confidence = 0

if contamination < 0.1:
    confidence += 0.4

if planet_radius_rj < 2:
    confidence += 0.3

if best_depth < 0.02:
    confidence += 0.3

print(f"Planet confidence: {confidence * 100:.1f}%")
```
