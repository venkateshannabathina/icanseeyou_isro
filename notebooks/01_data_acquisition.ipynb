```python
import lightkurve as lk

search_result = lk.search_lightcurve(
    "TIC 25155310",
    mission="TESS"
)

search_result
```

```python
lc = search_result[0].download()

print(lc)
```

```python
lc.plot()
```

```python
lc_clean = lc.remove_nans().flatten()

lc_clean.plot()
```

```python
from astropy.timeseries import BoxLeastSquares
import numpy as np

time = lc_clean.time.value
flux = lc_clean.flux.value

model = BoxLeastSquares(time, flux)

periods = np.linspace(0.5, 10, 10000)

results = model.power(periods, 0.2)
```

```python
best_period = results.period[np.argmax(results.power)]

print(f"Detected period: {best_period:.3f} days")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(results.period, results.power)
plt.xlabel("Period (days)")
plt.ylabel("BLS Power")
plt.show()
```

```python
period = best_period
depth = results.depth[np.argmax(results.power)]
duration = results.duration[np.argmax(results.power)]
```

```python
print("Period:", period)
print("Depth:", depth)
print("Duration:", duration)
```
